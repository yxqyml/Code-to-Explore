# 高维立方体体积分布
一个N维立方体的体积可以通过其边长的N次方来计算。假设边长为L, 那么N维立方体的体积可以用下式计算  
$$V_{cube}=L^N$$
而一个N维球体的体积可以通过以下公式计算  
$$
V_{sphere}=\frac{R^N\pi^{N/2}}{\Gamma(N/2+1)}
$$
如果我们考虑一个边长为2R的高维立方体和它的内接球，那么它们的体积之比可以通过下面的公式计算：  
$$
\rho(N)=\frac{V_{sphere}}{V_{cube}}=\frac{\pi^{N/2}}{\Gamma(N/2+1)2^N}
$$

```python
%matplotlib ipympl
import matplotlib.pyplot as plt   
import numpy as np
from scipy.special import gamma
from matplotlib.animation import FuncAnimation
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
sns.set()
```

```python
def r(N):
    return np.pi**(N/2) /(2**N*gamma(N/2+1))
    
```

```python
N_val = np.arange(2, 500)
r_val = r(N_val)

fig, ax = plt.subplots()
ax.set_xlabel('N-dimension')
ax.set_ylabel('r(N)')

def update(frame):
    ax.cla()
    x = [i for i in range(frame)]
    y = [r for r in r_val[:frame]]
    ax.bar(x, y)

ani = FuncAnimation(fig, 
                    update, 
                    frames=np.arange(2, 500), 
                    interval=50
        )
plt.show()
```

$\rho$随着维度N的增加迅速减小。这意味着，在高维空间中，大部分的体积集中在立方体的角落，而不是在其内接球中。  这和我们能够直观感知到的二维和三维空间不同。
在二维空间中，正方形的面积和其内接圆的面积之比是一个经典的几何关系。考虑一个边长为 a 的正方体，它的面积是 $A_{square}=a^2$，而内接圆的半径为$\frac{a}{2}$​，因此其面积是 $A_{circle}=π(a/2)^2$。

那么，和内接圆和其正方形的面积之比为：
$$
\rho_{2d}=\frac{A_{circle}}{A_{square}}=\frac{\pi (a/2)^2}{a^2} = \frac{\pi}{4}\approx 78\%
$$


在三维空间中，对于一个边长为 a 的立方体和其内接球，体积之比可以通过类似的方法得到。立方体的体积是 $V_{cube}=a^3，而内接球的半径为 $\frac{a}{2}$​，因此其体积是 $V_{sphere}=\frac{4}{3}\pi(\frac{a}{2})^3$.

正方体和内接球的体积之比为：

$$
\rho_{3d}=\frac{V_{sphere}}{V_{cube}}=\frac{\frac{4}{3}\pi(\frac{a}{2})^3}{a^3}​=\frac{\pi}{6} \approx 52 \%
$$
